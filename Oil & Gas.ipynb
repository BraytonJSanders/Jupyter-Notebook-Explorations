## Select only data for 'Green River - Over Thrust' Basin
Primarily is a South West Wyoming Basin

```python
df = df_raw[df_raw['DI Basin'] == 'GREEN RIVER - OVERTHRUST'].copy()
df.reset_index(inplace = True, drop = True)

print('{:,} Observations'.format(df.shape[0]))
display(df.head())
```

```python
# Map 1 for Is Public Company or 0 for Not Public. 
    # Maybe will be interesting to look into public v not public wells in the future
df['Public Co'] = df['Operator Ticker'].map(lambda x: 0 if pd.isna(x) else 1)
```

```python
# Where are the wells located
print('Number of Observations for Colorado: {:,}'.format(len(df[df['State'] == 'CO'])))
print('Number of Observations for Wyoming: {:,}'.format(len(df[df['State'] == 'WY'])))
```

```python
# Look at unique values in various columns
def unique_sorted(column):
    return list(np.sort(column.unique()))

print(unique_sorted(df['Drill Type']))
print(unique_sorted(df['County/Parish']))
print(unique_sorted(df['Entity Type']))
print('Number of Operators : {}'.format(len(unique_sorted(df['Operator Company Name']))))
```

```python
# Drop unwanted columns
df = df.drop(columns = ['Operator Alias (Legacy)', 'Operator (Reported)', 'Operator Ticker',
                  'Well/Lease Name', 'DI Play', 'DI Subplay', 'Country', 'PGC Area', 'Producing Status'])
```

```python
print('Basin Average Oil Gravity: %.3f' % df['Oil Gravity'].mean())
print(df.groupby(['County/Parish']).mean()['Oil Gravity'])
```

```python
# Percent of total oil produced by each country in the basin
print(df.groupby(['County/Parish']).sum()['Total Oil'] / (df.groupby(['County/Parish']).sum()['Total Oil'].sum()) * 100)
```

```python
# Percent wells located in each country
print(df.groupby(['County/Parish']).count()['API/UWI'] / (df.groupby(['County/Parish']).count()['API/UWI'].sum()) * 100)
```
