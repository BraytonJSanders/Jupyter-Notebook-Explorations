This program uses an artificial recurrent network called Long Short Term Memory (LSTM)
    to prodict the closing stock price of TSLA (Tesla, Inc.) using the past 60 day stock price.

```python
# Import Libraries

import math
import pandas_datareader as pdr
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error
from keras.models import Sequential
from keras.layers import Dense, LSTM
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')
```

```python
# Import TSLA hisoric data
# I only want to look into the 'Closing' price column
ticker = 'TSLA'

df_raw = pdr.DataReader(ticker, 
                        data_source='yahoo')
df_raw.tail()
```

## Describe Data

```python
# Number of: rows, columns
df_raw.shape
```

```python
# Starting date
df_raw.index[0]
```

```python
# Simple Summary Stats
df_raw['Close'].describe()
```

```python
# Check for missing values
print(df_raw['Close'].isna().sum()) # NaN values
len(df_raw[df_raw['Close'] == 0]['Close']) # Maybe missing value is in as zero
```

```python
# View the stock data
plt.figure(figsize = (16,8))
plt.title(ticker + ' Close Price History')
plt.plot(df_raw['Close']);
plt.xlabel('Date', fontsize = 18);
plt.ylabel('Close Price in USD ($)', fontsize = 18);
plt.show();
```

```python
# Select on the the data I want to continue with
df = df_raw.filter(['Close'])
df.head()
```

```python
# Convert to NumPy array
dataset = df.values
```

```python
# Get number of row to train the data on
training_data_len = math.ceil(len(dataset) * 0.8)
training_data_len
```

```python
# Scale the data
scaler = MinMaxScaler(feature_range=(0,1))
scaled_data = scaler.fit_transform(dataset)
scaled_data[-5:]
```

```python
# Create the training data set
# Create the scaled training data set
train_data = scaled_data[:training_data_len, :]
```

```python
# Split data into x_train and y_train
x_train = []
y_train = []

for i in range(60, len(train_data)):
    x_train.append(train_data[i-60:i, 0])
    y_train.append(train_data[i, 0])
```

```python
# Convert to NumPy arrays
x_train, y_train = np.array(x_train), np.array(y_train)
```

```python
# Reshape the x_train (needs to be 3D for LSTM)
x_train = np.reshape(x_train, (x_train.shape[0], x_train.shape[1], 1))
x_train.shape
```

```python
# Build to LSTM model
model = Sequential()
model.add(LSTM(50, return_sequences = True, input_shape = (x_train.shape[1], x_train.shape[2])))
model.add(LSTM(50, return_sequences = False))
model.add(Dense(25))
model.add(Dense(1))
```

```python
# Compile the model
model.compile(optimizer = 'adam', loss = 'mean_squared_error')
```

```python
# train the model
model.fit(x_train, y_train, batch_size = 1, epochs = 1)
```

```python
# Create the testing data set
test_data = scaled_data[training_data_len - 60:, :]
```

```python
# Split data into x_test and y_test
x_test = []
y_test = dataset[training_data_len:, :]

for i in range(60, len(test_data)):
    x_test.append(test_data[i-60:i, 0])
    
x_test = np.array(x_test)
```

```python
# Reshape the x_train (needs to be 3D for LSTM)
x_test = np.reshape(x_test, (x_test.shape[0], x_test.shape[1], 1))
x_test.shape
```

```python
# Get the model's predicted values
predictions = model.predict(x_test)
predictions = scaler.inverse_transform(predictions)
```

```python
# get the model RMSE
rmse = np.sqrt(mean_squared_error(y_test, predictions))
print('RMSE: %.3f' % rmse)
```

```python
# Plot data all data
train = df[:training_data_len]
valid = df[training_data_len:]
valid['Predictions'] = predictions

plt.figure(figsize = (16,8))
plt.title('LSTM Model for ' + ticker + ' Close Price History')
plt.xlabel('Date', fontsize = 18);
plt.ylabel('Close Price in USD ($)', fontsize = 18);

plt.plot(train['Close']);
plt.plot(valid[['Close', 'Predictions']]);

plt.legend(['Train', 'Valid', 'Predicted'])
plt.show();
```

```python
# Plot data only valid and predicted prices
train = df[:training_data_len]
valid = df[training_data_len:]
valid['Predictions'] = predictions

plt.figure(figsize = (16,8))
plt.title('LSTM Model for ' + ticker + ' Close Price History')
plt.xlabel('Date', fontsize = 18);
plt.ylabel('Close Price in USD ($)', fontsize = 18);

plt.plot(valid[['Close', 'Predictions']]);

plt.legend(['Valid', 'Predicted'])
plt.show();
```

```python
# Show only valid and predicted prices
valid['% Diff'] = round((valid['Predictions'] - valid['Close']) / valid['Close'] *100,2)
valid.tail()
```

```python
# Predict tomorrow's price
last_60_days = df[-60:].values
last_60_days_scaled = scaler.transform(last_60_days)

x_test = []
x_test.append(last_60_days_scaled)
x_test = np.array(x_test)

x_test = np.reshape(x_test, (x_test.shape[0], x_test.shape[1], 1))
x_test.shape
```

```python
# Get new predicted prices
pred_price = model.predict(x_test)
pred_price = scaler.inverse_transform(pred_price)
pred_price = pred_price[0][0]

print('Predicted Price: $' + str(round(pred_price,2)))
```
