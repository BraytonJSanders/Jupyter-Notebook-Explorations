```python
import pickle

def load_CIFAR10(batch_list, test_file):
    Xtr_dummy = []
    Ytr_dummy = []
    for i, file in enumerate(batch_list):
        with open(file, 'rb') as fo:
            dict = pickle.load(fo, encoding='bytes')
            keys = list(dict.keys())
        Xtr_dummy.append(dict[keys[2]])
        Ytr_dummy.append(dict[keys[1]])
        
        if i == 1:
            Xtr = np.concatenate([Xtr_dummy[i-1], Xtr_dummy[i]], axis = 0)
            Ytr = np.concatenate([Ytr_dummy[i-1], Ytr_dummy[i]], axis = 0)
        elif (i != 0) & (i != 1):
            Xtr = np.concatenate([Xtr, Xtr_dummy[i]], axis = 0)
            Ytr = np.concatenate([Ytr, Ytr_dummy[i]], axis = 0)
            
    with open(test_file, 'rb') as fo:
        dict = pickle.load(fo, encoding='bytes')
        keys = list(dict.keys())
    Xte = np.array(dict[keys[2]])
    Yte = np.array(dict[keys[1]])
        
    return Xtr, Ytr, Xte, Yte
```
